```python
import pandas as pd
read=pd.read_csv('vt.csv')
print(read.info())
print(read)

print("\n\n\nFilna Mean\n\n")

mean=read['Calories'].mean()
print(mean)
read['Calories'].fillna(mean,inplace=True)
print(read)

print("\n\n\nChange data type\n")

read['Calories']=read['Calories'].astype('int')
print(read)
print("\n\n\nChange DateTime\n")

read['Date']=pd.to_datetime(read['Date'])
print(read)



print("\n\n\nDuplicate removing")
read.duplicated().any()
read.drop_duplicates(inplace=True)
print(read)


print("\n\n\nWrong Data\n")

```
